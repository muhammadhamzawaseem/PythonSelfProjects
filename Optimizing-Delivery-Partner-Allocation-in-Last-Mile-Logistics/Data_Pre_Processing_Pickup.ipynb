```python
# Split the 'accept_time' column by space into two columns: 'date' and 'time'
combined_df[['accept_date', 'accept_time1']] = combined_df['accept_time'].str.split(' ', expand=True)

# Drop the original 'accept_time' column
combined_df = combined_df.drop(columns=['accept_time'])

# Split the 'time_window_start' column by space into two columns: 'date' and 'time'
combined_df[['date_window_start', 'time_window_start1']] = combined_df['time_window_start'].str.split(' ', expand=True)

# Drop the original 'time_window_start'
combined_df = combined_df.drop(columns=['time_window_start'])

# Split the 'time_window_end' column by space into two columns: 'date' and 'time'
combined_df[['date_window_end', 'time_window_end1']] = combined_df['time_window_end'].str.split(' ', expand=True)

# Drop the original 'time_window_end' column 
combined_df = combined_df.drop(columns=['time_window_end'])

# Split the 'pickup_time' column by space into two columns: 'date' and 'time'
combined_df[['pickup_date', 'pickup_time1']] = combined_df['pickup_time'].str.split(' ', expand=True)

# Drop the original 'pickup_time' column
combined_df = combined_df.drop(columns=['pickup_time'])

# Split the 'time_window_start' column by space into two columns: 'date' and 'time'
combined_df[['date_pickup_gps', 'pickup_gps_time1']] = combined_df['pickup_gps_time'].str.split(' ', expand=True)

# Drop the original 'time_window_start'
combined_df = combined_df.drop(columns=['pickup_gps_time'])

# Split the 'time_window_end' column by space into two columns: 'date' and 'time'
combined_df[['date_accept_gps', 'accept_gps_time1']] = combined_df['accept_gps_time'].str.split(' ', expand=True)

# Drop the original 'time_window_end' column 
combined_df = combined_df.drop(columns=['accept_gps_time'])
```

```python
# Create a full date assuming the current year (e.g., 2024)
combined_df['accept_date'] = pd.to_datetime(combined_df['accept_date'] + '-2024', format='%m-%d-%Y')

# Convert 'accept_date' column to datetime and reformat
combined_df['accept_date'] = pd.to_datetime(combined_df['accept_date'], format='%m-%d-%Y').dt.strftime('%d-%m-%Y')

# Create a full date assuming the current year (e.g., 2024)
combined_df['date_window_start'] = pd.to_datetime(combined_df['date_window_start'] + '-2024', format='%m-%d-%Y')

# Convert 'accept_gps_date' column to datetime and reformat
combined_df['date_window_start'] = pd.to_datetime(combined_df['date_window_start'], format='%m-%d-%Y').dt.strftime('%d-%m-%Y')

# Create a full date assuming the current year (e.g., 2024)
combined_df['date_window_end'] = pd.to_datetime(combined_df['date_window_end'] + '-2024', format='%m-%d-%Y')

# Convert 'delivery_date' column to datetime and reformat
combined_df['date_window_end'] = pd.to_datetime(combined_df['date_window_end'], format='%m-%d-%Y').dt.strftime('%d-%m-%Y')

# Create a full date assuming the current year (e.g., 2024)
combined_df['pickup_date'] = pd.to_datetime(combined_df['pickup_date'] + '-2024', format='%m-%d-%Y')

# Convert 'delivery_gps_date' column to datetime and reformat
combined_df['pickup_date'] = pd.to_datetime(combined_df['pickup_date'], format='%m-%d-%Y').dt.strftime('%d-%m-%Y')

# Create a full date assuming the current year (e.g., 2024)
combined_df['date_pickup_gps'] = pd.to_datetime(combined_df['date_pickup_gps'] + '-2024', format='%m-%d-%Y')

# Convert 'delivery_date' column to datetime and reformat
combined_df['date_pickup_gps'] = pd.to_datetime(combined_df['date_pickup_gps'], format='%m-%d-%Y').dt.strftime('%d-%m-%Y')

# Create a full date assuming the current year (e.g., 2024)
combined_df['date_accept_gps'] = pd.to_datetime(combined_df['date_accept_gps'] + '-2024', format='%m-%d-%Y')

# Convert 'delivery_gps_date' column to datetime and reformat
combined_df['date_accept_gps'] = pd.to_datetime(combined_df['date_accept_gps'], format='%m-%d-%Y').dt.strftime('%d-%m-%Y')
```

```python
combined_df['accept_time'] = combined_df['accept_date'] + ' ' + combined_df['accept_time1']
combined_df['time_window_start'] = combined_df['date_window_start'] + ' ' + combined_df['time_window_start1']
combined_df['time_window_end'] = combined_df['date_window_end'] + ' ' + combined_df['time_window_end1']
combined_df['pickup_time'] = combined_df['pickup_date'] + ' ' + combined_df['pickup_time1']
combined_df['pickup_gps_time'] = combined_df['date_pickup_gps'] + ' ' + combined_df['pickup_gps_time1']
combined_df['accept_gps_time'] = combined_df['date_accept_gps'] + ' ' + combined_df['accept_gps_time1']
```

```python
combined_df = combined_df.drop(columns=['accept_time1'])
combined_df = combined_df.drop(columns=['time_window_start1'])
combined_df = combined_df.drop(columns=['time_window_end1'])
combined_df = combined_df.drop(columns=['pickup_time1'])
combined_df = combined_df.drop(columns=['pickup_gps_time1'])
combined_df = combined_df.drop(columns=['accept_gps_time1'])
combined_df = combined_df.drop(columns=['date_accept_gps'])
combined_df = combined_df.drop(columns=['date_pickup_gps'])
combined_df = combined_df.drop(columns=['pickup_date'])
combined_df = combined_df.drop(columns=['date_window_end'])
combined_df = combined_df.drop(columns=['date_window_start'])
combined_df = combined_df.drop(columns=['accept_date'])
```

```python
combined_df['accept_time'] = pd.to_datetime(combined_df['accept_time'], dayfirst=True)
combined_df['time_window_start']=pd.to_datetime(combined_df['time_window_start'], dayfirst=True)
combined_df['time_window_end'] = pd.to_datetime(combined_df['time_window_end'], dayfirst=True)
combined_df['pickup_time']=pd.to_datetime(combined_df['pickup_time'], dayfirst=True)
combined_df['pickup_gps_time'] = pd.to_datetime(combined_df['pickup_gps_time'], dayfirst=True)
combined_df['accept_gps_time']=pd.to_datetime(combined_df['accept_gps_time'], dayfirst=True)
combined_df['ds'] = combined_df['ds'].astype(int)
```

```python
#calculate time window duration in minutes
combined_df['time_window_duration'] = (combined_df['time_window_end'] - combined_df['time_window_start']).dt.total_seconds() / 60
```

```python
#calculate acceptance to pickup duration in minutes
combined_df['task_duration'] = (combined_df['pickup_time'] - combined_df['accept_time']).dt.total_seconds() / 60
```

```python
#calculate distance between acceptance and pickup using Euclidean
def euclidean_distance(lat1, lon1, lat2, lon2):
    # Calculate differences
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Calculate Euclidean distance
    distance = math.sqrt(dlat**2 + dlon**2)
    return distance

combined_df['distance'] = combined_df.apply(
    lambda row: euclidean_distance(row['accept_gps_lat'], row['accept_gps_lng'],
                                   row['pickup_gps_lat'], row['pickup_gps_lng']),
    axis=1
)
```

```python
Q1 = combined_df['task_duration'].quantile(0.25)  # First quartile (25th percentile)
Q3 = combined_df['task_duration'].quantile(0.75)  # Third quartile (75th percentile)

# Calculate the IQR
IQR = Q3 - Q1

# Define the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data within the bounds
filtered_df = combined_df[
    (combined_df['task_duration'] >= lower_bound) & 
    (combined_df['task_duration'] <= upper_bound)
]

print("Filtered Data:")
print(filtered_df[['task_duration']])

# Count the outliers removed
outliers_removed = len(combined_df) - len(filtered_df)
print(f"Number of outliers removed: {outliers_removed}")
```
